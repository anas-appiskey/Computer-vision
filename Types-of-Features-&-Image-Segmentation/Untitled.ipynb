```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
#READ in image
image = cv2.imread('images/white-hand.jpg')

#make a copy
image_copy = np.copy(image)

#convert to RGB
image_copy = cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB )

#plt image
plt.imshow(image_copy)
```
